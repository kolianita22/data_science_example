```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df=pd.read_csv('Downloads/clv_data.csv')
df

```

```python
df.head(10)
```

```python
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
z_scores = np.abs(stats.zscore(df.select_dtypes(include=[np.number])))
df_no_outliers = df[(z_scores < 3).all(axis=1)]  # Removing rows with outliers
print(f"Original dataset:{df.shape}")
print(df)
print(f"\nDataset after removing outliers:{df_no_outliers.shape}")
print(df_no_outliers)
```

```python
sns.boxplot(df['days_on_platform'],gap=0)
```

```python
def extract_outliers_from_boxplot(array):
    ## Get IQR
    q1 = np.quantile(array, 0.25)
    q3 = np.quantile(array, 0.75)
    med = np.median(array)

    # finding the iqr region
    iqr = q3-q1

    # finding upper and lower whiskers
    upper_bound = q3+(1.5*iqr)
    lower_bound = q1-(1.5*iqr)

    outliers = array[(array <= lower_bound) | (array >= upper_bound)]
    print('Outliers within the box plot are :{}'.format(outliers))
    return outliers

extract_outliers_from_boxplot(df['days_on_platform'])
```
