```python
import librosa
import matplotlib.pyplot as plt
```

```python
#wave the aduio file
waveform,sample_rate=librosa.load('BAK.wav')
#create a window 
plt.figure(figsize=(10,4))
#Draw the wavwform 
librosa.display.waveshow(waveform,sr=sample_rate,color="blue")
plt.title("Waveform ")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
waveform,sample_rate=librosa.load('BAK.wav')
#calculate the short-time Fourier Transform
stft=librosa.stft(waveform)

#convert the amplitude values to decbels
spectogram=librosa.amplitude_to_db(np.abs(stft))
#create a window 
plt.figure(figsize=(10,4))
#Draw the wavwform 
librosa.display.specshow(spectogram,sr=sample_rate,x_axis='time',y_axis='log')

#add a colorbar
plt.colorbar()

#set a ttile
plt.title("Spectogram (db)")

#Adjust layout
plt.tight_layout()

#display the plot
plt.show()
```
