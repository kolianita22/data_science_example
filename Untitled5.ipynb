```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python

# 1. Load and preprocess the dataset
# Load the 'tips' dataset from seaborn
df = sns.load_dataset('tips')
```

```python
# Display basic information
df.head(10)
```

```python
df.shape
```

```python
# Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
# 2. Detect and remove outliers using IQR method
def remove_outliers(df, columns):
    # Calculate the IQR for each specified column
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df
```

```python
# Remove outliers from numerical columns 'total_bill' and 'tip'
df_no_outliers = remove_outliers(df, ['total_bill', 'tip'])
```

```python
# Display the new dataset without outliers
print(f"\nDataset shape after removing outliers: {df_no_outliers.shape}")
```

```python
# 3. Maximum Likelihood Estimation (MLE) for Normal Distribution
mean_mle = df_no_outliers['total_bill'].mean()
std_mle = df_no_outliers['total_bill'].std()
```

```python
print(f"\nMaximum Likelihood Estimation for 'total_bill':")
print(f"Mean: {mean_mle}, Standard Deviation: {std_mle}")
```

```python
# 4. Data Visualization using Matplotlib and Seaborn
plt.figure(figsize=(12, 6))
```

```python
# Histogram for 'total_bill' with KDE
plt.subplot(1, 2, 1)
sns.histplot(df_no_outliers['total_bill'], kde=True, color='blue')
plt.title('Histogram and KDE for Total Bill')
```

```python
# 4.2 Boxplot using Seaborn
plt.subplot(1, 2, 2)
sns.boxplot(x='day', y='total_bill', data=df_no_outliers, palette='Set2')
plt.title('Boxplot for Total Bill by Day')

plt.tight_layout()
plt.show()
```
