```python
from nltk.tokenize import word_tokenize
from nltk.stem import SnowballStemmer
stemmer=SnowballStemmer(language='english')

word_to_stem=['running','happily','quickly']
stemmed_word=[stemmer.stem(word) for word in word_to_stem]

print(f'Original Word:{word_to_stemm}')
print("Stemmed  Word:",stemmed_word)
```

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
word_to_stem=['running','happily','quickly']
stemmed_word=[stemmer.stem(word) for word in word_to_stem]

print(f'Original Word:{word_to_stemm}')
print("Stemmed  Word:",stemmed_word)
```

```python
from nltk.stem import RegexpStemmer
custom_rule=r'ing$'
regex_stemmer=RegexpStemmer(custom_rule)
w='running'

stemmed_word=regex_stemmer.stem(w)
print(f'Original Word:{w}')
print("Stemmed  Word:",stemmed_word)
```

```python
from nltk.stem import LancasterStemmer
stemmer=LancasterStemmer()

word=['running','happily','quieckly']
stemmed_word=[stemmer.stem(w) for w in word]

print(f'Original Word:{word}')
print("Stemmed  Word:",stemmed_word)
```

```python
sentence="programmmer program with programming language,and they programesd"
words=word_tokenize(sentence)
stemmed_word=[stemmer.stem(w1) for w1 in words]
stemmed_sentence=" ".join(stemmed_word)

print(f'Original Word:{sentence}')
print("Stemmed  Word:",stemmed_word)
print("Sremmed sentence",stemmed_sentence)
```

```python
from nltk import pos_tag

text="Nltk is a powerful library for natural language processing."
w2=word_tokenize(text)
pos_tags=pos_tag(w2)

print("original Ttext",text)
print("\npos Tagging Result:")
for word1,pos_tag in pos_tags:
    print(f"{w2}:{pos_tags}")


```

```python
#lemmatiztion
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
text="the cats were runnning faster thsn dog"
token=word_tokenize(text)
lemmatized_word=[lemmatizer.lemmatize(word) for word in token]
print("original Text:",text)
print("Lemmatizer Word:",lemmatized_word)
```
