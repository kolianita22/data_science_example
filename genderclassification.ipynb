```python
import random

from nltk.corpus import names
import nltk
```

```python
def gender_features(word):
    return {'Last_letter':word[-1]}

label_name=([(name,'male') for name in names.words('male.txt')]+[(name,'female') for name in names.words('female.txt')])
```

```python
random.shuffle(label_name)
```

```python
featureset=[(gender_features(n),gender) 
          for(n,gender)in label_name]
```

```python
train_set,test_set=featureset[500:],featureset[:500]
```

```python
classifier=nltk.NaiveBayesClassifier.train(train_set)

print(classifier.classify(gender_features(input("Enter the Male and Female NAme"))))
```
