```python
Q1=data['y'].quantile(0.25)
Q3=data['y'].quantile(0.75)
IQR=Q3-Q1
lb=Q1-1.5*IQR
ub=Q3+1.5*IQR
IQR_outliers=((data['y']>=lb)|(data['y']<=ub))
```
