```python
import matplotlib.pyplot as plt
import librosa.display
import numpy as np
```

```python
audio_file='sample-3s.mp3'
y,sr=librosa.load(audio_file)
#calculating the mel spectogram of the audio
S=librosa.feature.melspectrogram(y=y,sr=sr)
#ploting mel spectogram
plt.figure(figsize=(10,4))
#displaying the melspectogram
librosa.display.specshow(librosa.power_to_db(S,ref=np.max),sr=sr,x_axis='time',y_axis='mel')
plt.title('mel spectogram')
plt.show()
```
