```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore
```

```python
# 1. Import data using Seaborn
data = sns.load_dataset('tips')

# 2. Export the data to a CSV file
data.to_csv('tips_data.csv', index=False)

# Display the first few rows of the dataset
print("Original Data:")
print(data.head(10))
```

```python
# 3. Detecting and Removing Outliers
# We'll use Z-score to detect outliers for numerical columns
numerical_columns = ['total_bill', 'tip', 'size']

# Calculate Z-scores
z_scores = np.abs(zscore(data[numerical_columns]))

# Set a threshold for outliers
threshold = 3
outliers = (z_scores > threshold)

# Remove rows with outliers
data_cleaned = data[~np.any(outliers, axis=1)]

print(f"Original data shape: {data.shape}")
print(f"Cleaned data shape (after removing outliers): {data_cleaned.shape}")
```

```python
# 4. Data Visualization
# a) Scatter plot
plt.figure(figsize=(5, 3))
sns.displot(data=data_cleaned, x='total_bill', y='tip', hue='sex')
plt.title('Scatter Plot: Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
# b) Box plot
plt.figure(figsize=(5, 3))
sns.catplot(data=data_cleaned, x='day', y='total_bill', palette='Set2')
plt.title('Box Plot: Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
plt.hist(data[numerical_columns[0]], bins=20, color='blue', alpha=0.7)
plt.title(f'Histogram of {numerical_columns[0]}')
plt.xlabel(numerical_columns[0])
plt.ylabel('Frequency')
plt.show()
```
