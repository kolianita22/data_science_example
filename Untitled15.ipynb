```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Load Seaborn's 'tips' dataset
data = sns.load_dataset('tips')
```

```python
data.head(10)
```

```python
# 1. Preprocess the dataset
# Introduce some missing values for demonstration (optional)
data.loc[0:5, 'total_bill'] = np.nan
print("Missing values before preprocessing:")
print(data.isnull().sum())

# Fill missing values with the median of each column (for demonstration)
data['total_bill'].fillna(data['total_bill'].median(), inplace=True)

# Check missing values after preprocessing
print("\nMissing values after preprocessing:")
print(data.isnull().sum())
```

```python
# 2. Detect and remove outliers using Z-score method
numerical_columns = ['total_bill', 'tip', 'size']
z_scores = np.abs(zscore(data[numerical_columns]))  # Calculate Z-scores for numerical columns

# Define a threshold for outlier detection (usually 3)
threshold = 3
outliers = (z_scores > threshold)  # Identify outliers based on the Z-score threshold

# Display rows that are outliers
print("\nRows identified as outliers based on Z-score threshold:")
print(data[np.any(outliers, axis=1)])

# Remove outliers from the dataset
data_cleaned = data[~np.any(outliers, axis=1)]

# Check the shape of the cleaned dataset
print("\nShape of cleaned dataset (after removing outliers):", data_cleaned.shape)
```

```python

# 3. Frequency Distribution: Display the distribution of 'total_bill' using a histogram
plt.figure(figsize=(5, 3))
plt.hist(data_cleaned['total_bill'], bins=20, color='skyblue', edgecolor='black')
plt.title('Frequency Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
# 4. Data Visualization using Matplotlib

# a) Box Plot: Display the distribution of 'total_bill' by 'day' to compare
plt.figure(figsize=(5, 3))
sns.boxplot(x='day', y='total_bill', data=data_cleaned, palette='Set2')
plt.title('Box Plot: Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
# b) Scatter Plot: Relationship between 'total_bill' and 'tip'
plt.figure(figsize=(5, 3))
plt.scatter(data_cleaned['total_bill'], data_cleaned['tip'], color='green', alpha=0.5)
plt.title('Scatter Plot: Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()


```
