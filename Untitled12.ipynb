```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore
```

```python
# Load Seaborn 'tips' dataset
data = sns.load_dataset('tips')

# Introduce some missing values for demonstration (optional)
data.loc[0:5, 'total_bill'] = np.nan

# 1. Preprocess the Dataset
# Check and handle missing data (if any)
print("Missing values before preprocessing:")
print(data.isnull().sum())
```

```python


# Check missing values after preprocessing
print("\nMissing values after preprocessing:")
print(data.isnull().sum())
```

```python
# 2. Implement and Compare Outlier Detection Methods
# (a) Z-score Method
numerical_columns = ['total_bill', 'tip', 'size']
z_scores = np.abs(zscore(data[numerical_columns]))  # Calculate Z-scores

# Set a threshold for Z-score to detect outliers
threshold = 3
z_outliers = (z_scores > threshold)

# Remove rows where any of the Z-scores are above the threshold
data_z_cleaned = data[~np.any(z_outliers, axis=1)]
```

```python
# (b) IQR Method (Interquartile Range)
Q1 = data[numerical_columns].quantile(0.25)
Q3 = data[numerical_columns].quantile(0.75)
IQR = Q3 - Q1

# Define the outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detect outliers using the IQR method
IQR_outliers = ((data[numerical_columns] < lower_bound) | (data[numerical_columns] > upper_bound))

# Remove rows where any value is outside the IQR bounds
data_IQR_cleaned = data[~np.any(IQR_outliers, axis=1)]
```

```python
# Compare the shapes of the original and cleaned datasets
print(f"\nOriginal data shape: {data.shape}")
print(f"Data shape after Z-score outlier removal: {data_z_cleaned.shape}")
print(f"Data shape after IQR outlier removal: {data_IQR_cleaned.shape}")
```

```python
# 3. Data Visualization using Seaborn

# a) Scatter Plot: Relationship between 'total_bill' and 'tip'
plt.figure(figsize=(5, 5))
sns.scatterplot(data=data_z_cleaned, x='total_bill', y='tip', hue='sex', palette='coolwarm')
plt.title('Scatter Plot: Total Bill vs Tip (Z-score cleaned data)')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
# b) Box Plot: Distribution of 'total_bill' across days
plt.figure(figsize=(5, 5))
sns.boxplot(data=data_IQR_cleaned, x='day', y='total_bill', palette='Set2')
plt.title('Box Plot: Total Bill by Day (IQR cleaned data)')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```
